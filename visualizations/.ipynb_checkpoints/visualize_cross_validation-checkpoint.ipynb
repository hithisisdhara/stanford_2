## loss plot

```python
import matplotlib.pyplot as plt
def loss_plot(df):
    plt.plot(range(len(df['epoch'])), df['train_loss'].tolist(), 'go-')
    plt.plot(range(len(df['epoch'])), df['validation_loss'].tolist(), 'bo-')
    plt.plot(range(len(df['epoch'])), df['test_loss'].tolist(), 'ro-')
    #plt.title('model accuracy')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train','validation', 'test'],loc ='lower left')
    return plt
def acc_plot(df,plt):
    plt.plot(range(len(df['epoch'])), df['train_acc'].tolist(), 'go-')
    plt.plot(range(len(df['epoch'])), df['validation_acc'].tolist(), 'bo-')
    plt.plot(range(len(df['epoch'])), df['test_acc'].tolist(), 'ro-')
    #plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train','validation', 'test'], loc='lower right')
    return plt
```

```python
import pandas as pd
df = pd.read_csv('2deep_crossval_data.csv')

plt = loss_plot(df)
plt.show()
```

```python
plt = acc_plot(df,plt)
plt.show()
```
