```python
f = open('train_test.txt')
train,test = [],[]
for line in f:
    tr,te = [float(x) for x in line.split(",")]
    train.append(tr)
    test.append(te)
f.close()
```

```python
import matplotlib.pyplot as plt
plt.plot(range(len(train)), train, 'ro--')
plt.plot(range(len(test)), test, 'bo-')
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')
plt.show()
```
